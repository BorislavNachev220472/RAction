# Setup Dataset Assets

```python
import os
import shutil

def setup_dataset_assets(current_image_folder: str, current_label_folder: str,
                         new_image_folder: str, new_label_folder: str, labels_all: list):
    if not os.path.exists(new_image_tr) or not os.path.exists(new_label_tr):
        os.mkdir(new_image_tr)
        os.mkdir(new_label_tr)
        print('Output folders successfully created!')

    for _, folder, files in os.walk(current_image_folder):
        for file in files:
            corresponding_label = file[:4]
            if corresponding_label in labels_all:
                shutil.copy(src=os.path.join(current_image_folder, file),
                            dst=os.path.join(new_image_folder, file))

                current_label_name = f'{labels_all[labels_all.index(corresponding_label)]}.nii.gz'
                shutil.copy(src=os.path.join(current_label_folder, current_label_name),
                            dst=os.path.join(new_label_folder, current_label_name))
```

```python
data_prefix = '/workspace/2024-25ab-fai3-specialisation-project-team-specifix-1/data/segmentation'

current_image_tr = data_prefix + '/imagesTr'
current_label_tr = data_prefix +'/labels_h2/p1DU1'

dataset_path = 'Dataset061_p1DU1_h2'

os.mkdir(data_prefix + f'/nnUNet_raw/{dataset_path}')

new_image_tr = data_prefix + f'/nnUNet_raw/{dataset_path}/imagesTr'
new_label_tr = data_prefix + f'/nnUNet_raw/{dataset_path}/labelsTr'
correst_classes_list_m1 = list(map(lambda x: x[:4],os.listdir(current_label_tr)))

setup_dataset_assets(current_image_tr, current_label_tr, new_image_tr, new_label_tr, correst_classes_list_m1)
```
